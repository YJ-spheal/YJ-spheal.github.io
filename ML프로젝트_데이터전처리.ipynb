```python
import pandas as pd
K_income = pd.read_csv("Kor_income_training_data.csv")
K_income.head()
```

```python
K_income.info()
```

```python
elements = K_income[['year','wave','region','family_member','gender','year_born','education_level','marriage','religion','occupation','company_size','reason_none_worker']]
income = K_income['income']
```

```python
elements.isnull().sum()
```

```python
elements['occupation'].replace(' ',0,inplace = True)
elements['company_size'].replace(' ',0,inplace = True)
elements['reason_none_worker'].replace(' ',0,inplace = True)

elements['occupation'] = elements['occupation'].apply(int)
elements['company_size'] = elements['company_size'].apply(int)
elements['reason_none_worker'] = elements['reason_none_worker'].apply(int)
income = income.apply(int)
```
