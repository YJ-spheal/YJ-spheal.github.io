```python
import pandas as pd
import csv
f = open('Kor_income_training_data.csv', 'r', encoding = 'cp949')
data = csv.reader(f,delimiter = ',')
data = list(data)
f.close()
```

```python
new_K = list()
a = 0
for row in data:
    if row[-4] == ' ' and  row[-3] == ' ' and  row[-2] == ' ':
        a += 1
        continue
    new_K.append(row)
print(a)
```

```python
df = pd.DataFrame(new_K)
df.to_csv('KITD_rmv.csv',index = False, header = False)
```
